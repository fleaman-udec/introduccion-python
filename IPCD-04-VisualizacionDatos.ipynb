# Gr치ficos simples

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.linspace(0, 100, 31)
print(a)
plt.plot(a)
plt.show()
```

```python
x = np.arange(100)*0.01
fun = np.cos(2*np.pi*x)
plt.plot(x, fun, color='red', linewidth=3)
plt.xlabel('tiempo s')
plt.ylabel('amplitud V', fontsize=20)
plt.grid()
plt.show()
```

```python
fun2 = np.cos(2*5*np.pi*x)
plt.plot(x, fun, color='red', label='1 Hz')
plt.plot(x, fun2, color='black', label='5 Hz')
plt.xlabel('tiempo s', fontsize=14)
plt.ylabel('amplitud V', fontsize=14)
plt.legend(fontsize=20, loc='upper right')
plt.show()
```

```python
a = np.random.rand(10)
b = np.random.rand(10)
plt.scatter(a, b, marker='p')
plt.title('Grafico de dispersion')
plt.grid()
plt.show()
```

```python
c = np.random.rand(10)
plt.scatter(a, b, marker='s', label='puntos ab')
plt.scatter(a, c, marker='o', label='puntos ac')
plt.title('Grafico de dispersion')
plt.legend()
plt.grid()
plt.show()
```

# Gr치ficos para 3 variables

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

xv, yv = np.meshgrid(x, y)
zv = np.sin(np.sqrt(xv**2 + yv**2))

#cmaps = viridis, plasma, Reds
fig, ax = plt.subplots()
heatmap = ax.pcolormesh(xv, yv, zv, shading='auto', cmap='viridis')
ax.set_xlabel('variable x'), ax.set_ylabel('variable y')
cbar = plt.colorbar(heatmap)
cbar.set_label('variable z')
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(xv, yv, zv, color='lightblue')
plt.show()
```

```python
#%matplotlib qt
%matplotlib inline
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(xv, yv, zv, cmap='coolwarm', linewidth=0)
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_zticks([])
plt.show()
```

```python
np.random.seed(88)
x1 = np.random.randn(100)*10
x2 = 50 + 10*np.random.randn(100)
x3 = 50 + 10*np.random.randn(100)
x4 = 5.5 + 1.5*x3 - 3.3*np.log2(x3) + 15*np.random.rand(100)
x5 = -12.3*np.log2(x3) + 15*np.random.rand(100)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x3, x4, x5, marker='s', color='red', label='set 1')
ax.scatter(x1, x2, x3, marker='o', color='blue', label='set 2')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
cont = ax.contourf(xv, yv, zv, cmap='plasma', levels=20)
ax.set_xlabel('variable x'), ax.set_ylabel('variable y')
cbar = plt.colorbar(cont)
cbar.set_label('variable z')
plt.show()
```
