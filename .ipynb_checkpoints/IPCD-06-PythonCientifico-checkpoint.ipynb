# Optimización e interpolación

```python
import numpy as np
from scipy.optimize import minimize

# Definimos una función cualquiera para minimizar
def objective_function(x):
    return x**2 + 5*np.sin(x)

# Estimacion inicial
x0 = 2.0

# MInimizacion
result = minimize(objective_function, x0, method='BFGS')

print("El valor óptimo de X es:", result.x)
print("El valor mínimo de la función para este X es:", result.fun)
```

```python
from scipy.optimize import root

# Definimos funcion para encontrar sus raices
def equation(x):
    return x**2 - 2*x + 1

# Estimacion inicial
x0 = 0

# Solucion
solution = root(equation, x0)

print("La raiz de la ecuacion es:", solution.x)
```

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Generamos algunos datos
x = np.linspace(0, 10, 10)
y = np.sin(x)

# Creamos funcion para interpolar
f = interp1d(x, y, kind='cubic')

# Obtenemos valores de la funcion en varios puntos
x_new = np.linspace(0, 10, 100)
y_new = f(x_new)

# Grafico
plt.plot(x, y, 'o', label='datos brutos')
plt.plot(x_new, y_new, '-', label='interpolacion')
plt.legend()
plt.show()
```

# Álgebra lineal

```python
from scipy.linalg import solve

# Matriz de coeficientes
A = np.array([[3, 1], [1, 2]])

# Vector de datos
b = np.array([9, 8])

# Solucion del sistema
x = solve(A, b)

print("Solucion:", x)
```

```python
from scipy.linalg import det, eig, inv, norm

print('Inverso de la matriz = ', inv(A))

print('\nDeterminante de la matriz = ', det(A))

print('\nValores y vectores propios de la matriz = ', eig(A))

print('\nNorma del vector = ', norm(b))
```

# Procesamiento de señales

```python
from scipy.fft import fft, ifft

# Creamos una señal de prueba
fs = 100
t = np.arange(0, 1, 1/fs)
x = 2*np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t) + 0.2 * np.random.randn(100)

# Calculamos la FFT con scipy
M = fft(x)
M = np.abs(M)/len(x)
M = M[0: int(len(x)/2) ] 
M[1:] = 2*M[1:]

f = np.array([i*fs/len(x) for i in range(len(M))])

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title('Señal original')

plt.subplot(2, 1, 2)
plt.plot(f, M)
plt.title('Espectro de frecuencias')

plt.show()
```

```python
from scipy.signal import butter, filtfilt

# Diseño de filtro
b, a = butter(3, 20/fs, 'lowpass') #orden, frecuencia, tipo

# Aplicar filtro
x_filt = filtfilt(b, a, x)

# Graficos
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title('Señal original')

plt.subplot(2, 1, 2)
plt.plot(t, x_filt)
plt.title('Señal filtrada')

plt.show()
```

# Estadística y probabilidades

```python
from scipy.stats import norm

# Parametros de una distribucion normal
mu, sigma = 0, 1

# Se crea objeto distribucion
distribution = norm(mu, sigma)

# Genera valores aleatorios de la distribucion
samples = distribution.rvs(size=1000)

# Grafico del histograma de los valores
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Grafico de la funcion de densidad
x = np.linspace(-4, 4, 1000)
pdf = distribution.pdf(x)
plt.plot(x, pdf, 'k', linewidth=2)

plt.title('Distribucion normal')
plt.show()
```

```python
from scipy.stats import linregress

# Datos
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.1, 4.5, 6.1, 8.5, 9.7])

# Calculo de regresion lineal
slope, intercept, r_value, p_value, std_err = linregress(x, y)

print("Pendiente:", slope)
print("Intersección:", intercept)
print("R2:", r_value**2)

# Grafico
plt.scatter(x, y, label='Datos brutos')
plt.plot(x, slope*x + intercept, 'r', label='Recta ajustada')
plt.legend()
plt.show()
```

```python
from scipy.stats import pearsonr

# Coeficiente correlacion y valor p
correlation, p_value = pearsonr(x, y)

print("Coeficiente correlacion:", correlation)
print("Valor P:", p_value)
```

```python
from scipy.stats import f_oneway

# Datos de prueba para ANOVA
group1 = np.array([85, 90, 88, 75, 95])
group2 = np.array([78, 85, 80, 82, 89])
group3 = np.array([92, 88, 91, 87, 93])


# ANOVA de 1 factor
f_statistic, p_value = f_oneway(group1, group2, group3)

print("Estadistico F:", f_statistic)
print("Valor P:", p_value)
```
